```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv(r"C:\Users\SAI\Downloads\IRIS - IRIS.csv")
df.head()
```

```python
x=df.iloc[:,0:4]
x
```

```python
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
```

```python
#Univariate
for i, feature in enumerate(features): 
    plt.subplots(figsize=(10,5))
    sns.scatterplot(x=df[feature],y=[0]*len(df), hue=df['species']) 
    plt.title(f'Univariate Analysis of {feature}') 
    plt.xlabel(feature) 
    plt.ylabel('Constant Value (0)')
plt.tight_layout()
plt.show()
```

```python
# Bivariate Analysis
bivariate_pairs = [ ('sepal_width', 'petal_width'), ('sepal_length', 'petal_length')]
```

```python
plt.figure(figsize=(10, 5))
for i, (x_feature, y_feature) in enumerate(bivariate_pairs): 
    plt.subplot(1, 2, i+1) 
    sns.scatterplot(data=df, x=x_feature, y=y_feature, hue='species') 
    plt.title(f'{x_feature} vs {y_feature}')
plt.tight_layout()
plt.show()
```
